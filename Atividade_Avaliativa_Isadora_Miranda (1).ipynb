**Questão 1 — Classificação de
Temperaturas (Condicionais + Loop)**

**Tarefa 1.** Criar uma lista com as classificações

```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]

for temp in temperaturas:
    if temp < 20:
        print(f"{temp} → Frio")
    elif 20 <= temp <= 30:
        print(f"{temp} → Agradável")
    else: # temp > 30
        print(f"{temp} → Quente")
```

**Tarefa 2.** Contar quantos dias houve em cada categoria

```python
classificacoes = []
frio_count = 0
agradavel_count = 0
quente_count = 0

for temp in temperaturas:
    if temp < 20:
        classificacao = "Frio"
        frio_count += 1
    elif 20 <= temp <= 30:
        classificacao = "Agradável"
        agradavel_count += 1
    else: # temp > 30
        classificacao = "Quente"
        quente_count += 1
    classificacoes.append(f"{temp} → {classificacao}")

print("Classificações:", classificacoes)
print(f"\nTotal de dias Frios: {frio_count}")
print(f"Total de dias Agradáveis: {agradavel_count}")
print(f"Total de dias Quentes: {quente_count}")
```

**Questão 2 — Sistema de Avaliação de
Alunos**

**Tarefa 1.** Gerar lista de classificações

```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

for nota in notas:
    if nota < 5:
        print(f"{nota} → Reprovado")
    elif 5 <= nota < 7:
        print(f"{nota} → Recuperação")
    else: # nota >= 7
        print(f"{nota} → Aprovado")
```

**Tarefa 2.1** Calcular quantidade de aprovados

```python
aprovados_count = 0

for nota in notas:
    if nota >= 7:
        aprovados_count += 1

print(f"Total de alunos Aprovados: {aprovados_count}")
```

**Tarefa 2.2** Calcular percentual de reprovados

```python
reprovados_count = 0
total_alunos = len(notas)

for nota in notas:
    if nota < 5:
        reprovados_count += 1

if total_alunos > 0:
    percentual_reprovados = (reprovados_count / total_alunos) * 100
    print(f"Percentual de alunos Reprovados: {percentual_reprovados:.2f}%")
else:
    print("Não há alunos para calcular o percentual de reprovados.")
```

**Questão 3 — Monitoramento de
Consumo de Energia**

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

print("Classificação do Consumo Diário:")
for daily_consumo in consumo:
    if daily_consumo < 100:
        print(f"{daily_consumo} kWh → Baixo")
    elif 100 <= daily_consumo <= 180:
        print(f"{daily_consumo} kWh → Moderado")
    else: # daily_consumo > 180
        print(f"{daily_consumo} kWh → Alto")
```

**Tarefa 1.** Filtrar dias com consumo alto

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

print("Dias com consumo Alto:")
for daily_consumo in consumo:
    if daily_consumo > 180:
        print(f"{daily_consumo} kWh")
```

**Tarefa 2.1** Calcular total de consumo da semana

```python
total_consumo = sum(consumo)
print(f"Total de consumo da semana: {total_consumo} kWh")
```

**Tarefa 2.2** Calcular média semanal

```python
media_semanal = total_consumo / len(consumo)
print(f"Média semanal de consumo: {media_semanal:.2f} kWh")
```

**Tarefa 3.** Exibir alerta se houver mais de 2 dias com consumo alto

```python
alto_count = 0
for daily_consumo in consumo:
    if daily_consumo > 180:
        alto_count += 1

if alto_count > 2:
    print(f"ALERTA: Há {alto_count} dias com consumo de energia ALTO esta semana!")
else:
    print(f"Não há alerta de consumo alto (apenas {alto_count} dias com consumo alto).")
```

**Questão 4 — Simulação de Carrinho de
Compras**

```python
precos = [50, 120, 30, 200, 80, 15]

print("Preços com Desconto:")
for preco in precos:
    if preco < 50:
        desconto = 0.05  # 5% de desconto
    elif 50 <= preco <= 150:
        desconto = 0.10  # 10% de desconto
    else:  # preco > 150
        desconto = 0.15  # 15% de desconto

    preco_com_desconto = preco * (1 - desconto)
    print(f"Original: R${preco:.2f} | Desconto: {desconto*100:.0f}% | Final: R${preco_com_desconto:.2f}")
```

**Tarefa 1.** Criar lista com preços finais

```python
precos_finais = []

for preco in precos:
    if preco < 50:
        desconto = 0.05  # 5% de desconto
    elif 50 <= preco <= 150:
        desconto = 0.10  # 10% de desconto
    else:  # preco > 150
        desconto = 0.15  # 15% de desconto

    preco_com_desconto = preco * (1 - desconto)
    precos_finais.append(preco_com_desconto)

print("Lista de Preços Finais:", precos_finais)
```

**Tarefa 2.1** Calcular total da compra com desconto

```python
total_compra = sum(precos_finais)
print(f"Total da compra com desconto: R${total_compra:.2f}")
```

**Tarefa 2.2** Calcular economia total obtida

```python
total_original = sum(precos)
economia_total = total_original - total_compra
print(f"Economia total obtida: R${economia_total:.2f}")
```
